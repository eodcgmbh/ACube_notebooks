This notebook shows the SSM product created by TU Wien. TU Wien Surface Soil Moisture product measures the soil moisture of the very top soil layer (0-5cm). The product derived by using the TU Wien Change Detection Method on a time-series of Sentinel-1 high-resolution images.

The product name in the ACube is `SSM_Sentinel_1`. It is composed of a single measurement `VV`.

Relevant links:
* https://austriandatacube.eodc.eu/xwiki/bin/view/2.%20Terminology/3.4.%20data%20processing%20levels/3.4.3.%20Sentinel-1%20products/
* https://austriandatacube.eodc.eu/xwiki/bin/view/2.%20Terminology/3.3.%20Datasets/3.4.1.%20Sentinel-1%20data/3.4.1.3.%20Products/3.4.1.3.5.%20Surface%20Soil%20Moisture/

## Exploring the product

The code below initialises the datacube and lists some metadata of the product, measurements and datasets. We can see that this product has two measurements and we can also the spatial resolution of 500 m. We also list the metadata fields that are available.

```python
import datacube
from pprint import pprint
%matplotlib inline
dc = datacube.Datacube(app='sample_ssm')
product = 'SSM_Sentinel_1'
```

```python
prodlist=dc.list_products().dropna(axis=1)
prodlist[prodlist['name'] == product]
```

```python
measlist = dc.list_measurements()
measlist.loc[measlist.index.get_level_values(0) == product]
```

```python
query = {
    'lat': (48.00, 48.5),
    'lon': (16.0, 16.5),
    'time': ('2019-05-01', '2019-05-02')
}
```

```python
for dataset in dc.find_datasets_lazy(product=product, **query):
    print(100*'*')
    print('METADATA FIELDS: ', sorted(list(dataset.metadata_doc.keys())))
    print(100*'*')
    print('METADATA')
    pprint(dataset.metadata_doc)
    print(100*'*')
```

## Querying and displaying the data

```python
data = dc.load(product=product, 
               output_crs='EPSG:32633', 
               resolution=(-10,10),
                **query)
data
```

```python
data.VV.plot.imshow(
    x=data.crs.dimensions[1], 
    y=data.crs.dimensions[0],
    col='time',
    size=10,
    col_wrap=2,
    cmap='coolwarm'
)
```
